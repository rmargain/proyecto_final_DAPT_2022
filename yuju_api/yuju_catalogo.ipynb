```python
from dateutil import parser
import os
import requests
import pandas as pd
import time
from google.oauth2 import service_account
from google.cloud import firestore
from google.cloud import secretmanager
```

```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="../amazon_api/valoreo-production-306e2cc1a1fb.json"
path_to_credentials_file = os.environ.get('GOOGLE_APPLICATION_CREDENTIALS')
credentials = service_account.Credentials.from_service_account_file(path_to_credentials_file)
```

```python
elements = ['yuju-token']
```

```python
client = secretmanager.SecretManagerServiceClient()
```

```python
d1={}
for element in elements:
    try:
        parent = client.secret_path('valoreo-production', element)
        res = client.list_secret_versions(request ={'parent':parent})
        for version in res:
            name = version.name
            break
        response = client.access_secret_version(request={"name": name})
        d1[element] = response.payload.data.decode('utf8')
    except:
        d1[element] = None
```

```python
headers = {
    "Accept": "application/json",
    "Authorization": f"Token {d1['yuju-token']}"
}

base_url = 'https://api.software.madkting.com/shops/'
```
